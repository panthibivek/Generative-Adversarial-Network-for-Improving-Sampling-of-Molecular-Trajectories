# Visualizing the energies for Benzene Molecues generated by Generative Adv. Network

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
dirname = "/home/panthibivek/thesis/GAN_pkg/config_files_orca/calculated_energies.txt"
energies = []
with open(dirname, 'r') as f:
    energies = [float(line.strip()) for line in f if line]
energies = np.array(energies) * 27.2114079527
print("output data size:", energies.shape)
```

```python
plt.figure(figsize=(7,7))
bins_ = np.arange(4.4000, 4.70, 0.001000)
plt.hist(energies, bins=bins_, align='left', rwidth=0.9)
plt.xlabel('Bins', fontsize=15)
plt.ylabel('Count', fontsize=15)
fig2 = plt.gcf()
plt.show()
fig2.savefig('generated_data_distribution.png', dpi = 300)
```
