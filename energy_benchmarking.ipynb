# Visualizing the energies for Benzene Molecues generated by Generative Adv. Network

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
dirname = "/home/panthibivek/thesis/GAN_pkg/config_files_orca/calculate_energies.txt"
energies = []
with open(dirname, 'r') as f:
    energies = [float(line.strip()) for line in f if line]
energies = np.array(energies)
print("output data size:", energies.shape)
```

```python
from scipy import stats
plt.figure(figsize=(7,7))
plt.hist(energies, bins=30, color="lightblue", align='left', rwidth=0.9)
plt.xlabel('Bins', fontsize=15)
plt.ylabel('Count', fontsize=15)
fig1 = plt.gcf()

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 30)
p = stats.norm.pdf(x, np.mean(energies), np.std(energies))
print("Mean: {}".format(np.mean(energies)))
print("Standard dev.: {}".format(np.std(energies)))
plt.plot(x, p, 'k', linewidth=1, color="gray")
plt.show()
fig1.savefig('test_data_distribution.png', dpi = 300)
```
